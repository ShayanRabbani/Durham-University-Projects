calculate_score : This function adds up scores and points as given

change_background : This function updates the bacground once the avatar crosses a block

```python
from medialib import *
initialize()


def Positioning(x,y,imageRow,imageColumn,image):
    Array_positioning[imageRow-1][imageColumn-1] = [x,y,image]

    

def Background_position(j,x0,y0):
        if j=="X":
            draw('MazeGame_imgs/X.png', x0,y0)
        elif j=="_":
            draw('MazeGame_imgs/_.png', x0,y0)
        elif j=="L":
            draw('MazeGame_imgs/L.png', x0,y0)
        elif j=="M":
            draw('MazeGame_imgs/M.png', x0,y0)
        elif j=="J":
            draw('MazeGame_imgs/J.png', x0,y0)
        elif j=="E":
            draw('MazeGame_imgs/E.png', x0,y0)
        elif j=="B":
            draw('MazeGame_imgs/B.png', x0,y0)
        elif j=="H":
            draw('MazeGame_imgs/H.png', x0,y0)


def updateColumn(i,y0,imageRow,x0,imageColumn,image_len):
    for j in i:
        imageColumn=imageColumn+1
        Background_position(j,x0,y0)
        if game_running!=True:
            Positioning(x0,y0,imageRow,imageColumn,j)
        x0=x0+image_len

    
def updateRow(background,imageColumn,y0,x0,imageRow,image_len):
    for i in background:
        imageRow=imageRow+1
        updateColumn(i,y0,imageRow,x0,imageColumn,image_len)
        y0=y0+image_len
        

def calculate_score(score):
    global CurrentScore
    CurrentScore += score
    rect(10,20,200,20,r=200,g=200,b=200)
    text(f"Current points: {CurrentScore:.2f}",10,20,16)
    if CurrentScore==20 or CurrentScore>20:
        global game_over
        game_over = True
        clear(200,200,200)
        draw('MazeGame_imgs/you_win.png', 100,50)
        show_text("You win. Plese click to exit",250,20,20)

        
def show_text(str,x,y,clr):
    text(str,x,y,clr)    


def Background_updated():
    text('Maze game',300,100,20)
    text('You are a lost hero, and have to find your way back home...',10,500,16)
    text('When you collect 20 points, you can get back to home!',10,520,16)
    text('To play: press a and d keys to move left and right',10,540,16)
    text('         press w and s keys to move upwards and downwards',10,560,16)
    text('         press q to quit the game',10,580,16)

    
    
def Brit_Air():
    global CurrentScore
    global illegal_input_BA
    if illegal_input_BA == False:
        clear(200,200,200)
        text(f"Current points: {CurrentScore:.2f}",10,20,16)
        text("Welcome on board with British Airways ! You have recieved 1 point!", 50,50,16)
        draw('MazeGame_imgs/I.png', 10,200)
        text('However, now British Airways has an employee work crisis.',10,500,16)
        text('If you help them to make the decision, you get extra points!',10,520,16)
        text('   h. Hire new staff',10,540,16)
        text('   i. Increase existing staff pay',10,560,16)
        text('Please make your decision by inputing "h" or "i"',10,580,16)
    userinput1=wait_key_press()
    if userinput1=='h':
        clear(200,200,200) 
        Background_updated()
        illegal_input_BA = False
        updateRow(background1,imageColumn,y0,x0,imageRow,image_len)
        show_text('Thank you for supporting hiring new staffs! You recieved extra 5 points!',50,50,16)
        calculate_score(5)
    elif userinput1=='i':
        clear(200,200,200) 
        Background_updated()
        illegal_input_BA = False
        updateRow(background1,imageColumn,y0,x0,imageRow,image_len)
        show_text('Thank you for supporting increase in staff pay! You recieved extra 4 points!',50,50,16)
        calculate_score(4)
    else:
        rect(10,50,700,20,r=200,g=200,b=200)
        text("please enter a valid key",10,50,16)
        illegal_input_BA = True
        Brit_Air()
        


def change_background(avatar_xpos, avatar_ypos, next_xpos, next_ypos):
    global game_over
    is_British_Airways = False
    if Array_positioning[next_xpos][next_ypos][2] == "L":
        text("Welcome on board with Lufthansa ! You have recieved 1 point!", 50,50,16)
        calculate_score(1)
    elif Array_positioning[next_xpos][next_ypos][2] =="J":
        text("Welcome on board with Jet2 ! You have recieved 1 point!", 50,50,16)
        calculate_score(1)
     
    elif Array_positioning[next_xpos][next_ypos][2] =="E":
        text("Welcome on board with EasyJet. You have recieved 1 point!", 50,50,16)
        calculate_score(1)
       
    elif Array_positioning[next_xpos][next_ypos][2] == "B":
        is_British_Airways = True
        calculate_score(1)
        if game_over==False:
            Brit_Air()
    if  game_over==False:
        background1[avatar_xpos][avatar_ypos]="_"
        background1[next_xpos][next_ypos]="H"
        Positioning(Array_positioning[avatar_xpos][avatar_ypos][0],Array_positioning[avatar_xpos][avatar_ypos][1],avatar_xpos+1,avatar_ypos+1,"_")
        Positioning(Array_positioning[next_xpos][next_ypos][0],Array_positioning[next_xpos][next_ypos][1],next_xpos+1,next_ypos+1,"H")
        if is_British_Airways:
            updateRow(background1,imageColumn,y0,x0,imageRow,image_len)    

row_no=0
column_no=0


def findIndex(str,row_no,column_no,userinput):
    global game_over
    elementFound=False
    for i in background1:
        for avatar in i:
            if avatar==str:
                column_no=i.index(str)
                elementFound=True
                break
        if elementFound==True:
            break
        else:
            row_no=row_no+1
    if userinput=='a':
        if Array_positioning[row_no][column_no-1][2] == "X":
            show_text("The avatar cannot go through a wall",10,50,16)
        elif Array_positioning[row_no][column_no-1][2] == "M":
            clear()
            text("The avatar has come across a monster! GAME OVER", 50,50,16)
            game_over=True
        else:
            Background_position("_",Array_positioning[row_no][column_no][0],Array_positioning[row_no][column_no][1])
            Background_position("H",Array_positioning[row_no][column_no-1][0],Array_positioning[row_no][column_no-1][1])
            change_background(row_no,column_no,row_no,column_no-1)
    elif userinput=='d':
        if Array_positioning[row_no][column_no+1][2] == "X":
            show_text("The avatar cannot go through a wall",10,50,16)
        elif Array_positioning[row_no][column_no+1][2] == "M":
            clear()
            text("The avatar has come across a monster! GAME OVER", 50,50,16)
            game_over=True
        else:
            Background_position("_",Array_positioning[row_no][column_no][0],Array_positioning[row_no][column_no][1])
            Background_position("H",Array_positioning[row_no][column_no+1][0],Array_positioning[row_no][column_no+1][1])
            change_background(row_no,column_no,row_no,column_no+1)
    elif userinput=='w':
        if Array_positioning[row_no-1][column_no][2] == "X":
            show_text("The avatar cannot go through a wall",10,50,16)
        elif Array_positioning[row_no-1][column_no][2] == "M":
            clear()
            text("The avatar has come across a monster! GAME OVER", 50,50,16)
            game_over=True
        else:
            Background_position("_",Array_positioning[row_no][column_no][0],Array_positioning[row_no][column_no][1])
            Background_position("H",Array_positioning[row_no-1][column_no][0],Array_positioning[row_no-1][column_no][1])
            change_background(row_no,column_no,row_no-1,column_no)
    elif userinput=='s':
        if Array_positioning[row_no+1][column_no][2] == "X":
            show_text("The avatar cannot go through a wall",10,50,16)
        elif Array_positioning[row_no+1][column_no][2] == "M":
            clear()
            text("The avatar has come across a monster! GAME OVER", 50,50,16)
            game_over=True
        else:
            Background_position("_",Array_positioning[row_no][column_no][0],Array_positioning[row_no][column_no][1])
            Background_position("H",Array_positioning[row_no+1][column_no][0],Array_positioning[row_no+1][column_no][1])
            change_background(row_no,column_no,row_no+1,column_no)



background1=[['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X'],
             ['X','H','_','L','_','_','M','_','_','E','_','_','J','_','_','_','_','X'],
             ['X','_','_','_','M','_','B','_','L','_','_','_','_','B','_','J','_','X'],
             ['X','_','_','B','_','_','_','_','_','J','_','_','E','_','L','_','_','X'],
             ['X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X','X']]



rows_array = len(background1)
columns_array = len(background1[0])

Array_positioning=[[ ['0' for col in range(3)] for col in range(columns_array)] for row in range(rows_array)]

x0=10
y0=200 
imageRow=0
imageColumn=0
image_len= 30

clear(200,200,200) ## background
Background_updated()

game_running = False
CurrentScore=0
updateRow(background1,imageColumn,y0,x0,imageRow,image_len)
calculate_score(0)
game_over = False
illegal_input_BA = False

game_running= True
while game_running:
    global game_over
    if game_over==False:
        userinput = wait_key_press()
        if userinput=='a':
            rect(10,50,700,20,r=200,g=200,b=200)
            findIndex('H',row_no,column_no,userinput)
        elif userinput=='d':
            rect(10,50,700,20,r=200,g=200,b=200)
            findIndex('H',row_no,column_no,userinput)
        elif userinput=='w':
            rect(10,50,700,20,r=200,g=200,b=200)
            findIndex('H',row_no,column_no,userinput)
        elif userinput=='s':
            rect(10,50,700,20,r=200,g=200,b=200)
            findIndex('H',row_no,column_no,userinput)
        elif userinput=='q':
            
            clear()
            show_text("You have quit the game. Plese click to exit",200,20,20)
            print("Game Over!!!")
            
            break
        else:
            rect(10,50,700,20,r=200,g=200,b=200)
            show_text("please enter a valid key",10,50,16)
    else:
        break

wait_mouse_leftclick()
all_done()
```
